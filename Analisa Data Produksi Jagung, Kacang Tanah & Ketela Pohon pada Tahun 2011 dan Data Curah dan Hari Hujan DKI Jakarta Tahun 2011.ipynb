### Analisis Data Luas Panen dan Produksi Jagung, Kacang Tanah & Ketela DKI Jakarta Tahun 2011

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statisticsi as stat
dataPalawija = pd.read_csv('palawijajkt.csv', delimiter = ',', encoding = "ISO-8859-1")

```

### Deskripsi Data
Data ini merupakan data yang berisikan tentang hasil produksi dan luas panen tanaman Jagung, Ketela Pohon dan Kacang pada tahun 2011 di DKI Jakarta

```python
dataPalawija
```

#### Pengelompokan berdasarkan masa tanam dari setiap tanaman

```python
masaTanam = dataPalawija.groupby('masa_tanam').mean()
```

```python
masaTanam
```

```python
plt.plot(masaTanam.luas_panen)
plt.plot(masaTanam.produksi)
plt.title("Grafik Perbandingan Luas Panen dengan produksi Tanaman Selama 1 Tahun")
plt.xlabel("Masa Tanam")
plt.ylabel("Luas dalam m2 \n produksi dalam ton")
plt.legend()
plt.show()
```

Grafik diatas menunjukan perbandingan antara luas panen dengan hasil produksi yang ada di setiap masa tanam

#### Pengelompokan berdasarkan  nama tanaman

```python
tanaman = dataPalawija.groupby('jenis_tanaman').agg({'produksi': np.mean, 'luas_panen': np.mean})
```

```python
tanaman
```

### Pengelompokan Berdasarkan Tanaman 

```python
jagung = dataPalawija[dataPalawija['jenis_tanaman'] == 'Jagung']
ketela = dataPalawija[dataPalawija['jenis_tanaman'] == 'Ketela Pohon']
kacang = dataPalawija[dataPalawija['jenis_tanaman'] == 'Kacang Tanah']
```

```python
ketela
```

```python
statKetela = ketela.groupby('tahun').agg({'produksi': np.mean, 'luas_panen': np.mean})
statKetela
```

```python
plt.plot(jagung.masa_tanam, jagung.produksi, label="Produksi Jagung")
plt.plot(kacang.masa_tanam, kacang.produksi, label="Produksi Kacang")
plt.plot(ketela.masa_tanam, ketela.produksi, label="Produksi Ketela")

plt.title("Produksi Jagung, Kacang Tanah & Ketela Pada setiap musim Tanam")
plt.xlabel("Musim Tanam")
plt.ylabel("Berat (Ton)")
plt.legend()
plt.rcParams["figure.figsize"] = [10,10]
plt.show()
```

```python
plt.plot(tanaman.produksi, label= "produksi")
plt.plot(tanaman.luas_panen, label = "luas panen")
plt.title("Grafik Perbandingan Luas Panen dengan produksi setiap Tanaman Selama 1 Tahun Berdasarkan Jenis Tanaman")
plt.xlabel("Jenis Tanaman")
plt.ylabel("Luas dalam m2 \n produksi dalam ton")
plt.legend()
plt.show()
```
