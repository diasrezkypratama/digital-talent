```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## One to Many Relationship


### 1. Report the account representative for each customer.


```python
run("""
SELECT customername, contactfirstname, contactlastname, employeenumber, firstname, lastname
FROM customers INNER JOIN  employees ON salesrepemployeenumber = employeenumber limit 0,10
""")
```

## Many to Many Relationship



### 1. List products sold by order date.





```python
run("""
   SELECT 
       p.productName,
       od.priceEach,
       o.orderDate
   FROM
       products p  INNER JOIN orderdetails od 
   ON 
       p.productCode = od.productCode
       
       INNER JOIN orders  o  
   ON 
       od.orderNumber = o.orderNumber
    LIMIT 0,10
   
""")
```

### 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck

```python
run("""
   SELECT 
       p.productName,
       o.orderDate
   FROM
       products p  INNER JOIN orderdetails od 
   ON 
       p.productCode = od.productCode
       
       INNER JOIN orders  o  
   ON 
       od.orderNumber = o.orderNumber
   WHERE
       p.productName = "1940s Ford truck"
   ORDER BY
       o.orderDate desc
       limit 0,10
   
""")
```

### 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than 25,000?

```python
run("""
  SELECT
    c.customerName,
    c.customerNumber, 
    SUM(od.quantityOrdered * od.priceEach) AS Value
FROM customers as c, orders as o, orderdetails as od 
WHERE c.customerNumber = o.customerNumber 
      AND o.orderNumber = od.orderNumber
      GROUP BY o.orderNumber
HAVING Value >25000
limit 0,10


   
   
  
   
 
    
""")
```

### 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
    SELECT productName, buyPrice, priceEach
    FROM products INNER JOIN orderdetails
    on products.productCode = orderdetails.productCode
    WHERE priceEach = 2 * buyPrice
    limit 0,16
""")
```

### 7. List the products ordered on a Monday.

```python
run("""
    SELECT p.productname, DAYNAME(o.orderDate) as hari FROM products p, orderdetails od, orders o
    where p.productCode = od.productCode AND od.orderNumber = o.orderNumber
    HAVING hari like "Monday"
    limit 0,15
    
    
""")
```

### 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
    SELECT COUNT(p.quantityInStock) as onHoldOrders FROM products p, orderdetails od, orders o
    where p.productCode = od.productCode AND od.orderNumber = o.orderNumber AND o.status = "On Hold"
""")
```
